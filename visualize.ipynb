```python
import xarray as xr
import numpy as np

df = xr.open_dataset("data/processed_data.nc")
df
```

```python
# Use altair to plot the data
import altair as alt
import altair_saver

o3_spo = df['Total_Ozone'].to_dataframe()
o3_spo['time'] = [np.datetime64(i) for i in o3_spo.index] # np.arange('1963-01-01', "2022-03-01", dtype='datetime64[M]')

fig = alt.Chart(o3_spo).mark_line().encode(
    x='time:T',
    y='Total_Ozone:Q',
    tooltip=['time:T', 'Total_Ozone:Q']
)

fig
```
